```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
df =sns.load_dataset('tips')


print('Executive summary ')
print('-' *25)
print('Total revenue')
print('-' * 25)
print()
total_revenue = df['total_bill'].sum()
average_tip = (df['tip'].sum() / df['total_bill'].sum())*100


print(f'Total revenue is: ${total_revenue}')
print(f'Average tip: ${average_tip:.1f}%')
print(f'Total transection: {len(df):,}')
print(f'Average part size: {df['size'].mean():.0f} people' )

print('-' *25)
print('Key performane indicator')
print('-' *25)
print('Weekend vs weekday')
print('Lunch vs Dinner')
print('-' *25)

weekend_mask = df['day'].isin(['Sat', 'Sun'])
weekend_revenue = df[weekend_mask] ['total_bill'].sum()
weekday_revenue = df[~weekend_mask] ['total_bill'].sum()
food_time = df['time'].value_counts()

print(f'Weekend revenue: ${weekend_revenue:.2f}')
print(f'Weekday revenue: ${weekday_revenue:.2f}')
print(f"lunch: {food_time['Lunch']}")
print(f"Dinner: {food_time['Dinner']}")

print()
print('Customer behavior insight')
print('-' * 30)

smoker_tip = df.groupby('smoker')['tip'].mean()
smoker_count = df['smoker'].value_counts()
gender_bill = df.groupby('sex')['total_bill'].mean()

print()
print('Average bill by gender')
print('-' *25)
print(f" -Male: ${gender_bill['Male']:.2f}")
print(f" -Female: ${gender_bill['Female']:.2f}")

print()
print('Average tip by smoker')
print()
print(f" -Smokers: ${smoker_tip['Yes']:.2f} with {smoker_count['Yes']} of people")
print(f" -Non smokers: ${smoker_tip['No']:.2f} with {smoker_count['No']} of people")

print()
print('-' *20)
print('Recomandation')
print('-'*20)

print('stratigcal recommandation')
print('1: Focous on marketing on weekend -76% higher value')
print('2: Friday has lowest triffic consider promation')
print('3: Smoking section genarate diffrence spending pattern')
print('4: Average part size suggests family-friend positions')

print()
print('-' *15)
print('Visual data')
print('_' *15)

print('Total Revenu - Day Distribution')
day_revenu = df.groupby( 'day' )[ 'total_bill' ].sum()
plt.figure(figsize = (10, 6))
plt.bar(day_revenu.index, day_revenu.values, color=['blue', 'green', 'yellow', 'red'])
plt.title("Total revenu by day")
plt.xlabel("Days of the week")
plt.ylabel("Revenue ($)")
plt.savefig('total_revenue_by_day.png', dpi=300, bbox_inches='tight')
plt.show()

print('Food time - Bill Distribution')
meal_time = df['time'].value_counts()
plt.figure(figsize=(8,6))
plt.pie(
	meal_time,
	labels= meal_time.index,
	autopct='%1.1f%%',
    colors=['skyblue', 'pink'],
    startangle=90)
plt.title("Bill Distribution: Lunch vs Dinner")
plt.savefig('bill_distribution_meal_time.png', dpi=300, bbox_inches='tight')
plt.show()

print('Tips by smoker section')
smoker_tips = df['smoker'].value_counts()
plt.figure(figsize=(8,6))
plt.pie(
	smoker_tips, 
	labels= smoker_tips.index,
	autopct='%1.1f%%',
    colors=['lightgreen', 'blue'],
    startangle=90)
plt.title("Smoker Distribution: Smokers vs Non-smokers")
plt.savefig('tips_by_smoker.png', dpi=300, bbox_inches='tight')
plt.show()
```
